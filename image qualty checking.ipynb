```python
uqi_list = []
for images in generated_images_jpg:
    real_random_mild = rand.choice(real_images_mild)
    real_random_mild_image = cv2.imread(f"{real_mild_path}/{real_random_mild}")
    real_random_moderate = rand.choice(real_images_moderate)
    real_random_moderate_image = cv2.imread(f"{real_moderate_path}/{real_random_moderate}")
    real_random_non = rand.choice(real_images_non)
    real_random_non_image = cv2.imread(f"{real_non_path}/{real_random_non}")
    real_random_verymild = rand.choice(real_images_verymild)
    real_random_verymild_image = cv2.imread(f"{real_verymild_path}/{real_random_verymild}")
    generated_image = cv2.imread(f"{generated_path}/{images}")
    uqi_mild = uqi(real_random_mild_image, generated_image)
    uqi_moderate = uqi(real_random_moderate_image, generated_image)
    uqi_non = uqi(real_random_non_image, generated_image)
    uqi_verymild = uqi(real_random_verymild_image, generated_image)
    uqi_avg = (uqi_mild+uqi_moderate+uqi_non+uqi_verymild)/4
    uqi_list.append(uqi_avg)
    
print(len(uqi_list))
print(min(uqi_list), max(uqi_list))
x = np.arange(320)
plt.margins(0)
plt.xlabel("Generated Images")
plt.ylabel("UQI")
plt.plot(x, uqi_list)
```
